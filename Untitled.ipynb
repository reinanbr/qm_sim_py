```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros da simulação
L = 10.0           # Extensão do intervalo [-L, L]
N = 1024           # Número de pontos no espaço
dx = 2 * L / N     # Passo espacial
dt = 0.01          # Passo temporal
t_max = 1.0        # Tempo total de simulação
sigma = 1.0        # Largura do pacote de onda gaussiano
k0 = 5.0           # Número de onda inicial (momento central)

# Vetores de posição e número de onda
x = np.linspace(-L, L, N)
k = 2 * np.pi * np.fft.fftfreq(N, d=dx)

# Pacote de onda inicial
x0 = -L / 2
psi_x = np.exp(-(x - x0)**2 / (2 * sigma**2)) * np.exp(1j * k0 * x)
psi_k = np.fft.fft(psi_x)

# Operador de evolução temporal no espaço de frequência
U_k = np.exp(-1j * 0.5 * k**2 * dt)

# Evolução temporal
t = 0
while t < t_max:
    psi_k = U_k * psi_k  # Aplica o operador de evolução
    psi_x = np.fft.ifft(psi_k)  # Transforma de volta para o espaço real
    t += dt

    # Visualização (opcional)
    plt.plot(x, np.abs(psi_x)**2, label=f't = {t:.2f}')
    plt.xlabel("x")
    plt.ylabel("|ψ(x)|^2")
    plt.ylim(0, 1)
    plt.pause(0.01)  # Atualiza a figura
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros da simulação
L = 10.0           # Extensão do intervalo [-L, L]
N = 1024           # Número de pontos no espaço
dx = 2 * L / N     # Passo espacial
dt = 0.01          # Passo temporal
t_max = 1.0        # Tempo total de simulação
sigma = 1.0        # Largura do pacote de onda gaussiano
k0 = 5.0           # Número de onda inicial (momento central)

# Vetores de posição e número de onda
x = np.linspace(-L, L, N)
k = 2 * np.pi * np.fft.fftfreq(N, d=dx)

# Pacote de onda inicial
x0 = -L / 2
psi_x = np.exp(-(x - x0)**2 / (2 * sigma**2)) * np.exp(1j * k0 * x)
psi_k = np.fft.fft(psi_x)

# Operador de evolução temporal no espaço de frequência
U_k = np.exp(-1j * 0.5 * k**2 * dt)

# Configuração da visualização
plt.ion()  # Modo interativo
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8))

# Evolução temporal
t = 0
while t < t_max:
    # Evolução no espaço de frequência
    psi_k = U_k * psi_k  # Aplica o operador de evolução
    psi_x = np.fft.ifft(psi_k)  # Transforma de volta para o espaço real
    
    # Atualização do tempo
    t += dt
    
    # Visualização
    ax1.clear()
    ax1.plot(x, np.real(psi_x), label="Re[ψ(x)]", color="blue")
    ax1.plot(x, np.imag(psi_x), label="Im[ψ(x)]", color="red")
    ax1.set_ylabel("ψ(x)")
    ax1.set_title(f"Função de onda ψ(x) e seu módulo |ψ(x)| em t = {t:.2f}")
    ax1.legend()
    
    ax2.clear()
    ax2.plot(x, np.abs(psi_x), label="|ψ(x)|", color="purple")
    ax2.set_xlabel("x")
    ax2.set_ylabel("|ψ(x)|")
    ax2.legend()
    
    plt.pause(0.01)  # Atualiza a figura

plt.ioff()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse import diags
from scipy.sparse.linalg import splu

# Parâmetros da simulação
L = 10.0          # Extensão do intervalo [-L, L]
N = 1024          # Número de pontos no espaço
dx = 2 * L / N    # Passo espacial
dt = 0.005        # Passo temporal
t_max = 1.0       # Tempo total de simulação
sigma = 0.5       # Largura do pacote de onda gaussiano
k0 = 4.0          # Número de onda inicial (momento central)
V0 = 5.0          # Altura do degrau de potencial

# Vetor de posição
x = np.linspace(-L, L, N)

# Potencial degrau
V = np.zeros(N)
V[x >= 0] = V0

# Pacote de onda inicial
x0 = -L / 2
psi_x = np.exp(-(x - x0)**2 / (2 * sigma**2)) * np.exp(1j * k0 * x)

# Construção da matriz de Crank-Nicolson
alpha = 1j * dt / (2 * dx**2)
beta = -1j * dt / 2 * V

# Elementos da matriz A e B
main_diag = 1 + 2 * alpha + beta
off_diag = -alpha * np.ones(N-1)

# Matrizes A e B
A = diags([main_diag, off_diag, off_diag], [0, -1, 1], format="csr")
B = diags([1 - 2 * alpha - beta, alpha * np.ones(N-1), alpha * np.ones(N-1)], [0, -1, 1], format="csr")

# Pré-fator LU para resolver sistemas lineares
A_inv = splu(A) 

# Configuração da visualização
plt.ion()  # Modo interativo
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8))

# Evolução temporal
t = 0
while t < t_max:
    # Atualiza a função de onda psi usando o método de Crank-Nicolson
    b = B @ psi_x  # Multiplica B por psi_x
    psi_x = A_inv.solve(b)  # Resolve A @ psi_x = b

    # Atualização do tempo
    t += dt
    
    # Visualização
    ax1.clear()
    ax1.plot(x, np.real(psi_x), label="Re[ψ(x)]", color="blue")
    ax1.plot(x, np.imag(psi_x), label="Im[ψ(x)]", color="red")
    ax1.set_ylabel("ψ(x)")
    ax1.set_title(f"Função de onda ψ(x) e seu módulo |ψ(x)| em t = {t:.2f}")
    ax1.legend()
    
    ax2.clear()
    ax2.plot(x, np.abs(psi_x), label="|ψ(x)|", color="purple")
    ax2.plot(x, V / V0, label="Potencial V(x)", color="green")  # Normaliza V para visualização
    ax2.set_xlabel("x")
    ax2.set_ylabel("|ψ(x)|")
    ax2.legend()
    
    plt.pause(0.01)  # Atualiza a figura

plt.ioff()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros da simulação
L = 10.0          # Extensão do intervalo [-L, L]
N = 1024          # Número de pontos no espaço
dx = 2 * L / N    # Passo espacial
sigma = 0.5       # Largura do pacote de onda gaussiano
k0 = 5.0          # Número de onda inicial (momento central)
Ek = k0**2 / 2    # Energia cinética do pacote de onda
V0 = 4.0          # Altura do potencial degrau (escolhida para comparação com Ek)

# Vetor de posição
x = np.linspace(-L, L, N)

# Potencial degrau
V = np.zeros(N)
V[x >= 0] = V0  # Definição do potencial degrau

# Pacote de onda inicial |ψ(x)|
x0 = -L / 2
psi_x_initial = np.exp(-(x - x0)**2 / (2 * sigma**2)) * np.exp(1j * k0 * x)
mod_psi_x_initial = np.abs(psi_x_initial)

# Plotando |ψ(x)| e o potencial degrau V(x)
plt.figure(figsize=(10, 6))
plt.plot(x, mod_psi_x_initial, label="|ψ(x)| inicial", color="purple")
plt.plot(x, V / V0, label="Potencial degrau (V(x))", color="green", linestyle="--")
plt.axhline(Ek, color='blue', linestyle=':', label=f"Energia do pacote (E_k = {Ek:.2f})")

# Ajustes do gráfico
plt.xlabel("x")
plt.ylabel("Amplitude")
plt.title("Distribuição Inicial do Pacote de Onda e Potencial Degrau")
plt.legend()
plt.ylim(0, 1.5)
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse import diags
from scipy.sparse.linalg import splu

# Parâmetros da simulação
L = 10.0          # Extensão do intervalo [-L, L]
N = 1024          # Número de pontos no espaço
dx = 2 * L / N    # Passo espacial
dt = 0.005        # Passo temporal
t_max = 1.0       # Tempo total de simulação
sigma = 0.5       # Largura do pacote de onda gaussiano
k0 = 2.0          # Número de onda inicial (momento central)
Ek = k0**2 / 2    # Energia cinética do pacote de onda
V0 = 4.0          # Altura do potencial degrau (menor que Ek para gerar reflexão parcial)

# Vetor de posição
x = np.linspace(-L, L, N)

# Potencial degrau
V = np.zeros(N)
V[x >= 0] = V0

# Pacote de onda inicial
x0 = -L / 2
psi_x = np.exp(-(x - x0)**2 / (2 * sigma**2)) * np.exp(1j * k0 * x)

# Construção da matriz de Crank-Nicolson
alpha = 1j * dt / (2 * dx**2)
beta = -1j * dt / 2 * V

# Elementos da matriz A e B
main_diag = 1 + 2 * alpha + beta
off_diag = -alpha * np.ones(N-1)

# Matrizes A e B
A = diags([main_diag, off_diag, off_diag], [0, -1, 1], format="csr")
B = diags([1 - 2 * alpha - beta, alpha * np.ones(N-1), alpha * np.ones(N-1)], [0, -1, 1], format="csr")

# Pré-fator LU para resolver sistemas lineares
A_inv = splu(A) 

# Configuração da visualização
plt.ion()  # Modo interativo
fig, ax = plt.subplots(figsize=(10, 6))

# Evolução temporal e plotagem da parte refletida
t = 0
while t < t_max:
    # Atualiza a função de onda psi usando o método de Crank-Nicolson
    b = B @ psi_x  # Multiplica B por psi_x
    psi_x = A_inv.solve(b)  # Resolve A @ psi_x = b

    # Atualização do tempo
    t += dt

    # Plotando a parte refletida
    mod_psi_x_reflected = np.where(x < 0, np.abs(psi_x), 0)  # Amplitude |ψ(x)| em x < 0 para a reflexão
    ax.clear()
    ax.plot(x, mod_psi_x_reflected, label="Parte refletida |ψ(x)|", color="orange")
    ax.plot(x, V / V0, label="Potencial degrau (V(x))", color="green", linestyle="--")
    ax.axhline(Ek, color='blue', linestyle=':', label=f"Energia do pacote (E_k = {Ek:.2f})")

    # Ajustes do gráfico
    ax.set_xlabel("x")
    ax.set_ylabel("Amplitude")
    ax.set_title(f"Parte Refletida do Pacote de Onda em t = {t:.2f}")
    ax.legend()
    ax.set_ylim(0, 1.5)
    plt.pause(0.01)  # Atualiza a figura em tempo real

plt.ioff()
plt.show()

```
